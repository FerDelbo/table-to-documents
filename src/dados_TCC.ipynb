````python
def calcular_metricas(retorno_sistema, documentos_relevantes, k=5):
    
    top_k_hits = retorno_sistema[0][:k]
    top_k_ids = [hit['corpus_id'] for hit in top_k_hits]
    
    
    acertos = 0
    for doc_id in top_k_ids:
        if doc_id in documentos_relevantes:
            acertos += 1
    # Calcular Precisão @ K
    # Formula: (Relevantes Recuperados) / K
    precision_at_k = acertos / k
    
    # Calcular Recall @ K
    # Formula: (Relevantes Recuperados) / (Total de Relevantes Existentes)
    total_relevantes = len(documentos_relevantes)
    
    if total_relevantes > 0:
        recall_at_k = acertos / total_relevantes
    else:
        recall_at_k = 0.0
        
    return precision_at_k, recall_at_k
````

````python
# ground_truth = {1,2,3,4,8}
ground_truth = {1,2,3,4}
p_5, r_5 = calcular_metricas(results, ground_truth, k=4)

print(f"Top 5 IDs retornados: {[x['corpus_id'] for x in results[0]]}")
print(f"IDs Relevantes esperados: {ground_truth}")
print("-" * 30)
print(f"Precision@5: {p_5:.2f} ({(p_5*100):.1f}%)")
print(f"Recall@5:    {r_5:.2f} ({(r_5*100):.1f}%)")
````
